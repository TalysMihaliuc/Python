<b>A fórmula da mágica de Joel Greenblatt:</b> Esta fórmula foi desenvolvida pelo investidor e professor de finanças Joel Greenblatt e é baseada na teoria de que ações com baixo preço e alto retorno sobre o capital investido (ROIC) tendem a ser boas escolhas de investimento. A fórmula é:
ROIC / (1 - Preço / Lucro)

Onde ROIC é o retorno sobre o capital investido e o preço / lucro é o preço da ação dividido pelo lucro por ação.

<b>A fórmula de Decio Bazin:</b> Esta fórmula foi desenvolvida pelo investidor e empresário Decio Bazin e é baseada na teoria de que ações com baixo preço e alto lucro por ação tendem a ser boas escolhas de investimento. A fórmula é:
Preço / Lucro x Lucro por Ação

Onde o preço / lucro é o preço da ação dividido pelo lucro por ação e o lucro por ação é o lucro líquido da empresa dividido pelo número de ações em circulação.

```python
#CRIANDO DF COM TODOS OS PROVENTOS
proventos = pd.DataFrame()
for acao in acoes : 
    params = {'papel': str(acao),'tipo': '2',}   
    response = requests.get('https://www.fundamentus.com.br/proventos.php', params=params, cookies=cookies, headers=headers)   
    if 'Nenhum provento encontrado' in response.text:
        continue
    else: 
        tips = pd.read_html(response.text)[0]
        tips['ACAO'] = acao
        proventos = pd.concat([proventos, tips], ignore_index=True)

proventos['Valor'] = proventos['Valor']/10000
```

# 3. Setores da bolsa
***

```python
#ACESSAR BIBLITECA FUNDAMENTOS CAPTURAR SETOR DAS AÇÕES
```

```python
#CRIAR LISTA DA AÇÕES
t = fundamentus.get_resultado_raw()
t.reset_index(inplace=True)
lista  = t.papel.tolist()
```

```python
proventos.head()
```

```python
#CONVERTER COLUNA DATA PARA DATETIME
proventos['Data'] = pd.to_datetime(proventos['Data'])
#CRIAR A COLUNA ANO
proventos['Ano'] = proventos['Data'].dt.year
```

```python
proventos.head() 
```

```python
#SELECINANDO PROVENTOS DO ULTIMOS 5 ANOS.
proventos_5_anos = proventos[proventos.Ano >= 2017]
```

```python
#AGREGANDO VALOR POR ANO
table = pd.pivot_table(proventos_5_anos, values='Valor', index=['ACAO'], columns=['Ano'], aggfunc=np.sum, fill_value=0)
```

```python
#CRIANDO DF
table  = pd.DataFrame(table).reset_index()
```

```python
#COLUNA DY_MEDIO DOS ULTIMOS 5 ANOS
table['DY_MEDIO'] = table[[2017,2018,2019,2020,2021,2022]].mean(axis=1)
```

```python
#FILTRA EMPRESAS QUE PAGARAM DIVIDENDOS EM TODOS OS ULTIMOS 5 ANOS
pagou_5_anos = table[ (table[2017] != 0) & 
                    (table[2018] != 0)& 
                    (table[2019] != 0)&
                    (table[2020] != 0)&
                    (table[2021] != 0)&
                    (table[2022] != 0)].reset_index(drop=True).sort_values(by='DY_MEDIO', ascending=False)
```

```python
#MERGE DOS DATAFRAMES
df_final_acoes_COM_DY = pd.merge(df_final_acoes,pagou_5_anos[['ACAO','DY_MEDIO']], left_on='ticker', right_on='ACAO', how='left')
```

```python
#DROPAR COLUNA REPETIDA, RETIRAR TODOS OS VALORES NaN
df_final_acoes_COM_DY = df_final_acoes_COM_DY.drop(columns=["ACAO",'index'])
df_final_acoes_COM_DY.fillna(0, inplace=True)
```

```python
df_final_acoes_COM_DY.shape
```

# 4 (Joel Greenblatt)
***
 Não queremos qualquer tipo de empresa, então vamo fazer o seguinte:
 * empresas que tem liquidez diária de mais de R$ 100.000
 * Tiveram crescimento de lucros nos ultimos 5 anos
 

```python
#Remover acoes com liquides menor que 100.000, 
#Filtra acoes  lucros crescente nos ultimos 5 anos 
#Remover setores de utilidade Publica
setores = ['Água e Saneamento','Energia Elétrica']
df_magic_formula = df_final_acoes_COM_DY[(df_final_acoes_COM_DY.liquidezMediaDiaria > 100000) 
                                         & (df_final_acoes_COM_DY.lucros_Cagr5 > 1) 
                                         & (~df_final_acoes_COM_DY.Setor.isin(setores))]
```

```python
# FAZENDO RANKINGS

#RANKING EV_EBIT
df_magic_formula  = df_magic_formula .sort_values("eV_Ebit").reset_index(drop=True)
df_magic_formula  = df_magic_formula.reset_index()
df_magic_formula  = df_magic_formula .rename(columns={"index": "RKN_EV_EBIT"})
```

```python
#RANKING ROIC
df_magic_formula  = df_magic_formula .sort_values("roic").reset_index(drop=True)
df_magic_formula  = df_magic_formula.reset_index()
df_magic_formula  = df_magic_formula .rename(columns={"index": "RKN_ROIC"})
```

```python
#RANKING DA MAGIC FORMULA
df_magic_formula['MAGIC_FORMULA'] = df_magic_formula[['RKN_ROIC','RKN_EV_EBIT']].sum(axis=1)
```

```python
#TOP 10 AÇÕES DA FÓRMULA
df_magic_formula_f = df_magic_formula.sort_values(by ='MAGIC_FORMULA').reset_index(drop=True).head(10)
```

```python
#ORGANIZACAO DE COLUNAS
df_magic_formula_f = df_magic_formula_f[['MAGIC_FORMULA','companyName', 'ticker', 'price', 'p_L',
       'p_VP', 'p_Ebit', 'p_Ativo', 'eV_Ebit', 'margemBruta', 'margemEbit',
       'margemLiquida', 'p_SR', 'p_CapitalGiro', 'p_AtivoCirculante',
       'giroAtivos', 'roe', 'roa', 'roic', 'dividaliquidaPatrimonioLiquido',
       'dividaLiquidaEbit', 'pl_Ativo', 'passivo_Ativo', 'liquidezCorrente',
       'peg_Ratio', 'vpa', 'lpa', 'valorMercado', 'dy', 'liquidezMediaDiaria',
       'receitas_Cagr5', 'lucros_Cagr5', 'Setor', 'Subsetor', 'DY_MEDIO']].reset_index()
```

```python
#RENOMEAR INDICE
TOP10_MAGIC_FORMULA = df_magic_formula_f .rename(columns={"index": "MAGIC_INDEX"}).drop(columns=["MAGIC_FORMULA"])
TOP10_MAGIC_FORMULA
```

# 5 (Decio Bazin turbo)
***
Método Decio Bazin turbo
* DY > 6%
* Endividamento divida bruta/PL < 1
* ou (divida liquida/ EBITDA) < 3
* ROE > 15%

```python
#SELEÇÃO CRITÉRIO DECIO BAZIN
df_decio_bazin = df_final_acoes_COM_DY[(df_final_acoes_COM_DY.dy > 6) 
                                         & (df_final_acoes_COM_DY.dividaliquidaPatrimonioLiquido < 1) 
                                         & (df_final_acoes_COM_DY.roe > 15)
                                         & (df_final_acoes_COM_DY.DY_MEDIO > 0)].sort_values(by='roe',ascending=False).reset_index(drop=True).head(10)
```

```python
#TOP 10 BAZIN INDEX
df_decio_bazin_top_10 = df_decio_bazin.reset_index()
df_decio_bazin_top_10 = df_decio_bazin_top_10.rename(columns={"index": "BAZIN_INDEX"})
df_decio_bazin_top_10
```
